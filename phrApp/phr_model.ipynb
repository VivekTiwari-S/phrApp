```python
import pandas as pd
```

```python
data = pd.read_csv('krups_phr_data.csv')
data.info()
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_with_label_encoder(dataframe, column_name='Product-Category'):
    le = LabelEncoder()
    dataframe[column_name] = le.fit_transform(dataframe[column_name])
    return dataframe
```

```python
data = encode_with_label_encoder(data)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop('Health_Rating', axis=1)
y = data['Health_Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
with open('health_api.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)
```
